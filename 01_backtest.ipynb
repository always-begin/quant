## CAGR Compound Annual Growth rate

특정 기간의 수익률을 유지한다고 했을 때의 수익률

누적 수익률`Accumulated return`을 연평균 수익률로 환산해서 구함

if more than 1 year
- 복리

else
- 단리

```python
def __annualize(rate, period):
    if period < 360:
        rate = ((rate-1) / period * 365) + 1
    elif period > 365:
        rate = rate ** (365 / period)
    else:
        rate = rate
    return round(rate, 4)

```
