### 간단 계산식
![picture 2](../images/44da0c1db11251492c45d84ba951f61bff95866a79b86183a3db5dd366341cec.png)  

```
RSI 계산 공식 :

RSI = RS / (1 + RS)

또는, 다음과 같이 구해도 결과는 동일하다.

RSI = AU / (AU + AD)

```
- AU : 일간 상승 폭의 평균
- AD : 일간 하락 폭의 평균
